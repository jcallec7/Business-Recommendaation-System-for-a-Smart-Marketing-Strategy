```python
import pandas as pd
import numpy as np
url = "UsuarioCategoria(1).csv"
# load dataset into Pandas DataFrame
dataR = pd.read_csv(url, names=['Usuario','Categoria','Cantidad'])
dataR['Cantidad'] = dataR['Cantidad'].astype(int)
dataR
```

```python
matrizVentas = pd.pivot_table(dataR,index="Usuario",values=['Cantidad'],columns="Categoria")
print("Tabla Usuarios x Categorias")
dataset=matrizVentas.fillna(0)
dataset
```

```python
import math
from scipy import sparse
#from sklearn.preprocessing import StandardScaler
x=dataset
x = sparse.csr_matrix(x)
#x = StandardScaler().fit_transform(x)
print("Tabla Dispersa")
print(x)
```

```python
from sklearn.decomposition import TruncatedSVD
from sklearn.random_projection import sparse_random_matrix
num_components=10
svd = TruncatedSVD(n_components=num_components, n_iter=100, random_state=5)
principalComponents=svd.fit_transform(x)
print(svd.explained_variance_ratio_.sum()) 
```

```python
principalDf = pd.DataFrame(data = principalComponents)
principalDf
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import datasets

from sklearn.cluster import KMeans
 
def doKmeans(X, nclust=70):
    model = KMeans(nclust)
    model.fit(X)
    clust_labels = model.predict(X)
    cent = model.cluster_centers_
    return (clust_labels, cent)

clust_labels, cent = doKmeans(principalDf, 70)
kmeans = pd.DataFrame(clust_labels)

grupos = np.array(kmeans)
grupos
```

```python
kmeans = pd.DataFrame(grupos,columns=['Grupos'])
#kmeans.head()

UserGrupoK=kmeans.groupby(kmeans.Grupos).Grupos.count()
UserGrupoKS=UserGrupoK.sort_values(ascending=False)[:20]
grupo=UserGrupoKS.index.tolist()
valores=UserGrupoKS.values.tolist()

plt.rcParams.update({'font.size': 15})
plt.bar(range(len(grupo)), valores, color='orange')
##Pone el valor de los indices en el eje x
plt.xticks(range(len(grupo)), grupo, rotation='vertical')
plt.title("K-Means Groups")
plt.xlabel("Group Number")
plt.ylabel("Quantity of Customers")
plt.show()
```
